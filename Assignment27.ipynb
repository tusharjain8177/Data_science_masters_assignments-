**1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**

**Ans:** Seaborn is a Python library that is built on top of Matplotlib and is used for data visualization. It provides a higher-level interface than Matplotlib and can be used to create a variety of plots. Here are five plots that can be plotted using the Seaborn library along with their uses:

1. Histogram: A histogram is a plot that shows the distribution of a dataset by grouping data into bins along the x-axis and displaying the frequency or count of data in each bin along the y-axis. Seaborn provides the distplot() function to plot histograms with optional kernel density estimates (KDE) and rug plots.

2. Bar plot: A bar plot is a plot that shows the relationship between a categorical variable and a numeric variable. Seaborn provides the barplot() function to plot bar plots with optional error bars and confidence intervals.

3. Box plot: A box plot is a plot that shows the distribution of a dataset by displaying the quartiles, median, and outliers. Seaborn provides the boxplot() function to plot box plots with optional grouping by a categorical variable.

4. Heatmap: A heatmap is a plot that shows the relationship between two variables using color. Seaborn provides the heatmap() function to plot heatmaps with optional row and column normalization.

5. Line plot: A line plot is a plot that shows the relationship between two variables using a line. Seaborn provides the lineplot() function to plot line plots with optional grouping by a categorical variable and error bars.

**2. Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

fmri_data = sns.load_dataset('fmri')
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
```

**3. Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.**

Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic_data = sns.load_dataset('titanic')
fig, axs = plt.subplots(ncols=2, figsize=(10, 5))
sns.boxplot(x='pclass', y='age', data=titanic_data, ax=axs[0])
sns.boxplot(x='pclass', y='fare', data=titanic_data, ax=axs[1])
plt.show()
```

**4. Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds_data = sns.load_dataset('diamonds')
# diamonds_data

sns.histplot(data=diamonds_data, x='price', hue='cut')
```

**5. Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris_data = sns.load_dataset('iris')

sns.pairplot(iris_data, hue='species')
```
