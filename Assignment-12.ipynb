3. What errors are defined in the ArithmeticError class? Explain any two with an example.

Ans: The ArithmeticError class in Python is a base class for exceptions that occur during arithmetic operations. It has several subclasses, each of which represents a specific type of arithmetic error. Two examples of ArithmeticError subclasses are ZeroDivisionError and OverflowError.

- ZeroDivisionError: This error is raised when attempting to divide a number by zero. For example:

```python
x = 5
y = 0
z = x/y  # raises ZeroDivisionError
print(z)
```

- OverflowError: This error is raised when the result of an arithmetic operation is too large to be represented by the data type. For example:

```python
x = 2**1000
y = 2**1000
z = x*y  # raises OverflowError
z
```

4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans: The LookupError class in Python is a base class for exceptions that occur when a key or index is not found in a container object. It has several subclasses, each of which represents a specific type of lookup error. Two examples of LookupError subclasses are KeyError and IndexError.

- KeyError: This error is raised when a dictionary key is not found in the dictionary. For example:

```python
d = {'a': 1, 'b': 2, 'c': 3}
value = d['d']  # raises KeyError
print(value)
```

- IndexError: This error is raised when a list index is out of range or a string index is not found. For example:

```python
l = [1, 2, 3]
value = l[3]  # raises IndexError
print(value)
```

5. Explain ImportError. What is ModuleNotFoundError?

Ans: ImportError is a built-in exception class in Python that is raised when an imported module, function, or attribute cannot be found. This can occur for several reasons, such as a misspelled module name, a missing module in the Python path, or a missing dependent module.

ModuleNotFoundError is a subclass of ImportError that was introduced in Python 3.6. It is specifically raised when an imported module is not found in the Python path. This provides more specific information about the cause of the error and allows for more targeted handling of the error.
